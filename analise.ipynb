6. Graficos de Faturamento Por Loja

```python
loja.head(3)
```

```python
fat1 = loja['Preço'].sum()
fat2 = loja2['Preço'].sum()
fat3 = loja3['Preço'].sum()
fat4 = loja4['Preço'].sum()

faturamentos = [fat1, fat2, fat3, fat4]

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
plt.bar(lojas, faturamentos)

plt.title('Faturamento das Lojas')
plt.xlabel('Lojas')
plt.ylabel('Faturamento (R$)')
plt.xticks(rotation=0)

plt.show()
```

10. Grafico do frete medio.

```python
fretes = pd.Series({
    'Loja 1': loja['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
})

fretes = fretes.sort_values(ascending=True)
plt.barh(fretes.index, fretes.values)
plt.title('Frete médio por lojas')
plt.show()
```

Mesmo com o valor do frete sendo menor que as outras lojas, e as avalizaçoes serem iguais as da loja 1; o faturamento bruto da loja 4 é baixo comparado com as outras. Então, a loja que eu recomendo que seja vendida é a loja 4.
